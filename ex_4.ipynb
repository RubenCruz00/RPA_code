```python
import requests
from bs4 import BeautifulSoup
# URL de la página web de noticias deportivas "Marca"
url = 'https://www.marca.com/'
# Realizamos una solicitud GET para obtener el contenido de la página
response = requests.get(url)
# Verificamos si la solicitud fue exitosa
if response.status_code == 200:
    # Parseamos el contenido HTML de la página
    soup = BeautifulSoup(response.text, 'html.parser')
    # Buscamos las noticias relacionadas con "Topuria"
    noticias = soup.find_all('a', string=lambda text: text and 'Topuria'in text)
    # Imprimimos los titulares y enlaces de las noticias encontradas
if noticias:
    print("Noticias relacionadas con Topuria:")
    for noticia in noticias:
        print("- Título:", noticia.text)
        print(" Enlace:", noticia['href'])
    else:
        print("No se encontraron noticias relacionadas con Topuria.")
else:
    print("Error al obtener el contenido de la página:",response.status_code)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Inicializar el controlador de Selenium para Chrome
driver = webdriver.Chrome()

# Maximizar la ventana del navegador
driver.maximize_window()

# Navegar a la página de YouTube
driver.get("https://www.youtube.com/")

# Esperar a que cargue la página base
try:
    WebDriverWait(driver, 15).until(
        EC.presence_of_element_located((By.TAG_NAME, "ytd-app"))
    )
except:
    pass

# Intentar aceptar cookies si el botón aparece
try:
    accept_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="content"]/div[2]/div[6]/div[1]/ytd-button-renderer[2]/yt-button-shape'))
    )
    accept_button.click()
except:
    pass

# Esperar y hacer clic en el botón "En directo"
try:
    live_chip = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, '//yt-chip-cloud-chip-renderer//span[contains(text(),"En directo")]'))
    )
    live_chip.click()
except Exception as e:
    print("❌ No se encontró el botón 'En directo':", e)
    driver.quit()
    exit()

# Esperar a que se carguen los vídeos
WebDriverWait(driver, 20).until(
    EC.presence_of_element_located((By.ID, "video-title"))
)

# Obtener los títulos de los 5 primeros vídeos destacados
featured_videos = driver.find_elements(By.XPATH, '//*[@id="video-title"]')[:5]
for index, video in enumerate(featured_videos):
    print(f"Vídeo {index + 1}: {video.text}")

# Cerrar el navegador
driver.quit()
```

```python
import pyautogui
import time
import os

# 1. Abrir Word desde Spotlight (Command + Space)
pyautogui.hotkey('command', 'space')
time.sleep(1)
pyautogui.write('Word')
time.sleep(1)
pyautogui.press('enter')
time.sleep(10)  # esperar que Word abra
pyautogui.press('enter')

# 2. Escribir un texto
pyautogui.write('Este es un documento creado automáticamente con PyAutoGUI en macOS.')
time.sleep(4)

# 3. Guardar el documento usando la barra superior
# Acceso al guardado: Command + Shift + S (Guardar como)
pyautogui.hotkey('command', 'shift', 's')
time.sleep(10)

# Escribir nombre y seleccionar Escritorio como ubicación
pyautogui.write('Documento_Automatico.docx')
time.sleep(1)

# Navegar al escritorio
pyautogui.hotkey('command', 'shift', 'd')  # Ir al Escritorio
time.sleep(1)

# Confirmar guardado
pyautogui.press('enter')
time.sleep(2)

# 4. Cerrar Word (Command + Q)
pyautogui.hotkey('command', 'q')
time.sleep(2)

# 5. Mostrar el escritorio usando AppleScript
os.system('osascript -e \'tell application "System Events" to key code 103 using {command down}\'')

# Esperar a que el escritorio esté visible
time.sleep(2)

# 6. Capturar pantalla
screenshot = pyautogui.screenshot()
screenshot.save('captura_escritorio_mac.png')
print("✅ Captura guardada como 'captura_escritorio_mac.png'")
```
